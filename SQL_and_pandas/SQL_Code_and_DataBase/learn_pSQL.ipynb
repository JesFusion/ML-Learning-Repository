```python
import subprocess 
from jesse_custom_code.pandas_file import postgre_connect
from sqlalchemy import create_engine, text



"""
Checks if the Postgres Daemon is alive on your Ubuntu machine.
This mimics checking if the 'Bank Manager' is awake before calling him.
"""

print("===================================== Checking Postgres Status =====================================")

try:

    p_status = subprocess.check_output(["service", "postgresql", "status"]) # this runs a bash command from python to check service status of the postgreSQL server

    print('''
Postgres Service is Running...
    ''')

except subprocess.CalledProcessError:

    print('''
Postgres Service is down!
    ''')



print("\n===================================== Initiating Database Handshake =====================================\n")

try:

    database_engine = create_engine(postgre_connect)

    with database_engine.connect() as Dconnection: # The .connect() method is the actual handshake

        # let's run a simple query to verify is the Daemon hears us
        postgres_result = Dconnection.execute(text("SELECT version();"))

        print(f"Successful Handshake!\nConnected to: {postgres_result.fetchone()[0]}")

    print(database_engine)

except Exception as an_error:

    print(f"\nFailed Handshake!\nBank Manager Hung up. Here's the error: {an_error}\n")

    print(None)
```
